Cleaning Feb-2021 to Jun-2021:
* Load csv
* Drop duplicates and NA
* Change start and stop times to datetime
* Add tripduration column
* Rename columns to match previous spreadsheets
* Add season column
* Order columns

```python
Jun2021 = pd.read_csv('JC-202106-citibike-tripdata.csv')
df = pd.DataFrame(Jun2021)
df = df.drop_duplicates()
df = df.dropna(how='any')
df['ended_at'] = pd.to_datetime(df['ended_at'])
df['started_at'] = pd.to_datetime(df['started_at'])
df["tripduration"] = df["ended_at"] - df["started_at"]
df = df.rename(columns={"member_casual": "usertype", "started_at": "starttime", "ended_at": "stoptime"})
df["season"] = "Summer"
column_names = ["season", "starttime", "stoptime", "tripduration", "start_station_name", "start_station_id", "start_lat", "start_lng", "end_station_name",
                "end_station_id", "end_lat", "end_lng", "usertype", "rideable_type", "ride_id"]
Jun2021 = df.reindex(columns=column_names)
```
